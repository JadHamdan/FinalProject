```python
#import statements
import numpy as np # forlinear algebra
import matplotlib.pyplot as plt #for plotting things
import os
from PIL import Image
from sklearn.metrics import confusion_matrix
import seaborn as sns
import keras
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.preprocessing.image import ImageDataGenerator, load_img
```

```python
#creating model and adding the layers
model = Sequential()
model.add(Conv2D(32, (3, 3), activation="relu", input_shape=(64, 64, 3)))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Conv2D(32, (3, 3), activation="relu"))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Flatten())
model.add(Dense(activation = 'relu', units = 128))
model.add(Dense(activation = 'sigmoid', units = 1))
#compiling the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#data preprocessing
datagen = ImageDataGenerator(rescale=1./255)

batch_size = 32
train_generator = datagen.flow_from_directory(
        '/Users/jadhamdan/FinalProject/cell_images/train',  
        target_size=(64, 64), 
        batch_size=batch_size,
        class_mode='binary')
validation_generator = datagen.flow_from_directory(
        '/Users/jadhamdan/FinalProject/cell_images/validation',
        target_size=(64, 64),
        batch_size=batch_size,
        class_mode='binary')
test_generator = datagen.flow_from_directory(
        '/Users/jadhamdan/FinalProject/cell_images/test',
        target_size=(64, 64),
        batch_size=32,
        shuffle = False,
        class_mode='binary')
```

```python
x_batch, y_batch = next(train_generator)
for i in range (0,10):
    image = x_batch[i]
    print(y_batch[i])
    plt.imshow(image)
    plt.show()
```

```python
trained_model = model.fit_generator(train_generator,
                         steps_per_epoch = 163,
                         epochs = 10,
                         validation_data = validation_generator,
                         validation_steps = 624)
```

```python
model.save_weights('prelim.h5')
```

```python
test_accu = model.evaluate_generator(test_generator,steps=624)
print('The testing accuracy is :',test_accu[1]*100, '%')
```

```python
y_true = test_generator.classes
print(y_true)
```

```python
predictions = model.predict_generator(test_generator, steps=32, verbose=0)
y_pred = np.rint(predictions)
print(y_pred)
```

```python
mat = confusion_matrix(y_true, y_pred)

sns.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value');

plt.show()
```
